```python
import pandas as pd
```

```python
# Lendo um arquivo Excel com várias abas
excel_file = './data/Clientes.xlsx'
```

```python
# Sheet name = nome da planilha
todas_as_planilhas = pd.read_excel(excel_file, sheet_name=None)  # None significa todas as abas

todas_as_planilhas
```

```python
# Molezinha! Só passar o nome do arquivo!
# Mesmo se não passar sheet, ele lê e pega a primeira planilha
primeira_planilha = pd.read_excel(excel_file)

primeira_planilha
```

```python
# Lendo uma única aba
primeira_planilha = pd.read_excel(
    excel_file,
    sheet_name='Ceará',
)

primeira_planilha
```

```python
# Lendo uma única aba e pulando as duas primeiras linhas
primeira_planilha = pd.read_excel(
    excel_file,
    sheet_name='Ceará',
    skiprows=2
)

# skip = pular, rows = linhas

primeira_planilha
```

```python
# Lendo uma aba específica, usando apenas algumas colunas e definindo a primeira coluna como índice
somente_tres_colunas = pd.read_excel(
    excel_file,
    sheet_name='São Paulo',
    usecols='A:C',  # Pode usar um range de colunas ou uma lista de nomes de colunas
)

somente_tres_colunas
```

```python
# Lendo uma aba específica, usando apenas algumas colunas e definindo a primeira coluna como índice
somente_tres_colunas = pd.read_excel(
    excel_file,
    sheet_name='São Paulo',
    usecols='A:C',  # Pode usar um range de colunas ou uma lista de nomes de colunas
    index_col=0  # Define a primeira coluna como índice
)

somente_tres_colunas
```

```python
# Lendo um arquivo Excel, avaliando os tipos que foram inferifos pelo pandas
antes_do_dtype = pd.read_excel(
    excel_file,
    sheet_name='Santa Catarina',
)

antes_do_dtype.info()
```

```python
# Lendo um arquivo Excel, convertendo tipos de dados
tipos_float = pd.read_excel(
    excel_file,
    sheet_name='Santa Catarina',
    dtype={'ClienteID': float, 'Idade': float},  # Converte Coluna1 para string e Coluna2 para float
)

tipos_float
```
