## DataFrames Pandas - Filtros loc e iloc

Vamos compreender como funcionam os filtros com loc e iloc

```python
import pandas as pd
```

### Criando um DataFrame com índices e colunas nomeadas

```python
df = pd.DataFrame(data=[['costela', 30.], ['camarão', 60.]], 
                  index=['linha1', 'linha2'], 
                  columns=['prato', 'preço'])
df
```

### Importando um DataFrame da Web

```python
df_medalhas = pd.read_csv("https://raw.githubusercontent.com/letsdata/arquivos-jcd/main/medals.csv")

df_medalhas.head()
```

```python
df_medalhas.Sport
```

```python
df_medalhas['Sport']
```

```python
# loc NAO É POSICIONAL!!! ELE ACESSA O ÍNDICE
df.loc['linha1']
```

```python
# iloc É POSICIONAL!!!
df.iloc[0]
```

```python
df.iloc[1]
```

```python
# seu_dataframe.loc[seu_dataframe[sua_coluna_para_filtro] == 'criterio']
df_medalhas.loc[df_medalhas['NOC'] == 'FRA']
```

```python
# seu_dataframe.loc[(seu_dataframe[sua_coluna_para_filtro_1] == 'criterio_1') & 
# (seu_dataframe[sua_coluna_para_filtro_2] == 'criterio_2')]
df_medalhas.loc[(df_medalhas['NOC'] == 'FRA') & (df_medalhas['Medal'] == 'Gold')]
```

```python
# Saber se valor está na lista
df_medalhas.loc[(df_medalhas['NOC'] == 'FRA') & (df_medalhas['Medal'].isin(['Gold', 'Silver']))]
```

```python
# Saber se valor NÃO está na lista
df_medalhas.loc[(df_medalhas['NOC'] == 'FRA') & (~df_medalhas['Medal'].isin(['Gold', 'Silver']))]
```

```python
# seu_dataframe.loc[seus filtros de linha, [coluna1, coluna3, coluna7]]
df_medalhas.loc[(df_medalhas['NOC'] == 'FRA') & (df_medalhas['Medal'] == 'Gold'), ['Year', 'Sport', 'Event']]
```

```python
# Pipe (|) significa OU. O e-comercial (&) é o E (AND)
df_medalhas.loc[(df_medalhas['NOC'] == 'FRA') | (df_medalhas['NOC'] == 'AUT')]
```

```python
# Esse de cima é igual a esse de baixo
df_medalhas.loc[df_medalhas['NOC'].isin(['FRA', 'AUT'])]
```

```python
df_medalhas.columns
```

```python
# seleciona POSICIONALMENTE as linhas e as colunas
df_medalhas.iloc[0:3, 1:3]
```
